```python
import pandas as pd
```

```python
df = pd.read_csv("04_single_vs_multi_function_test_73.csv")
df
```

```python
smiles_list = df['SMILES'].tolist()
smiles_list
len(smiles_list)
```

```python
#Write the elements of the list to a text file
with open('smiles.txt', 'w') as f:
    for item in smiles_list:
        f.write("%s\n" % item)

print("Data has been written to the text file.")
```

```python
ligand = iFeatureOmegaCLI.iLigand("./smiles.txt")
```

```python
feature_type = ['EState', 'MACCS fingerprints', 'Morgan-ECFP4 fingerprints', 'Morgan-ECFP6 fingerprints', 'Morgan-FCFP4 fingerprints', 'Morgan-FCFP6 fingerprints']
```

```python
for f in feature_type:
    ligand.get_descriptor(f)
    encodings_df = pd.DataFrame(ligand.encodings)
    encodings_df.reset_index(drop=True, inplace=True)
    merged_df = pd.concat([df['Sample ID'], encodings_df, df['Label']], axis=1)
    csv_filename = f"04_single_vs_multi_function_test_73_ifeatureomega_{f}.csv"
    merged_df.to_csv(csv_filename, index=False)
    print(f"CSV file saved: {csv_filename}")
```
